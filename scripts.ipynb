```python
# Classes: Dealing with Complex Numbers

class Complex(object):
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary

        

    def __add__(self, no):
        c = Complex(self.real + no.real, self.imaginary + no.imaginary)
        return c

        

    def __sub__(self, no):
        c = Complex(self.real - no.real, self.imaginary - no.imaginary)
        return c

        

    def __mul__(self, no):
        c = Complex(self.real*no.real-(self.imaginary*no.imaginary), self.imaginary*no.real+self.real*no.imaginary)
        return c

        

    def __truediv__(self, no):
        divisor = no.real**2 + no.imaginary**2
        c = Complex((self.real*no.real+self.imaginary*no.imaginary)/divisor, (self.imaginary*no.real-self.real*no.imaginary)/divisor)
        return c

        

    def mod(self):
        c = Complex(math.sqrt(self.real**2+self.imaginary**2), 0)
        return c



    def __str__(self):

        if self.imaginary == 0:
            result = "%.2f+0.00i" % (self.real)

        elif self.real == 0:
            if self.imaginary >= 0:
                result = "0.00+%.2fi" % (self.imaginary)
            else:
                result = "0.00-%.2fi" % (abs(self.imaginary))
        elif self.imaginary > 0:
            result = "%.2f+%.2fi" % (self.real, self.imaginary)
        else:
            result = "%.2f-%.2fi" % (self.real, abs(self.imaginary))

        return result
```

```python
# Text Wrap

def wrap(string, max_width):
    new = ""
    for i in range(len(string)):
        if i *max_width+max_width < len(string):
            new += string[i*max_width:i *max_width+max_width]+"\n"
        else:
            new += string[i*max_width:]+"\n"

    return new
```

```python
# String Formatting
def print_formatted(number):

    # Determine the width of the binary representation of 'number'
    width = len(bin(number)) - 2

    for i in range(1, number + 1):
        # Convert the number to the required formats
        dec = str(i).rjust(width)
        octal = oct(i)[2:].rjust(width)
        hexadec = hex(i)[2:].upper().rjust(width)
        binary = bin(i)[2:].rjust(width)
        # Print the formatted string
        print(f"{dec} {octal} {hexadec} {binary}")
```

```python
# Piling Up!
# Enter your code here. Read input from STDIN. Print output to STDOUT


def can_pile_up(cubes):
    while len(cubes) > 1:
        if cubes[0] >= cubes[-1]:
            top = cubes.pop(0)
        else:
            top = cubes.pop()
        if len(cubes) > 0 and top < max(cubes[0], cubes[-1]):
            return "No"

    return "Yes"


T = int(input())


for _ in range(T):
    n = int(input())
    cubes = list(map(int, input().split()))
    print(can_pile_up(cubes))
```

```python
# Re.findall() & Re.finditer()
# Enter your code here. Read input from STDIN. Print output to STDOUT

def find_vowel_substrings(s):

    vowels = set("AEIOUaeiou")
    consonants = set("QWRTYPSDFGHJKLZXCVBNMqwrtypsdfghjklzxcvbnm")
    result = []
    n = len(s)
    i = 0

    while i < n:
        # Find the start consonant
        if s[i] in consonants:
            j = i + 1
            while j < n:
                if s[j] in consonants:
                    break
                j += 1

            if j < n and s[j] in consonants:
                # Check the substring between i and j
                substring = s[i+1:j]
                vowel_count = sum(1 for c in substring if c in vowels)
                if vowel_count >= 2:
                    result.append(substring)
            i = j
        else:
            i += 1

    if result:
        for res in result:
            print(res)
    else:
        print(-1)

find_vowel_substrings(input())
```

```python
# Hex Color Code
# Enter your code here. Read input from STDIN. Print output to STDOUT

import re



def extract_hex_codes(N, css_lines):

    hex_color_pattern = re.compile(r'#(?:[0-9a-fA-F]{3}|[0-9a-fA-F]{6})\b')
    valid_hex_codes = []
    in_brackets = False

    for line in css_lines:

        if in_brackets:
            matches = hex_color_pattern.findall(line)
            valid_hex_codes.extend(matches)

        if '}' in line:
            in_brackets = False
        if '{' in line:
            in_brackets = True


    for code in valid_hex_codes:
        print(code)

N = int(input())
css = []
for _ in range(N):
    css.append(input())

extract_hex_codes(N,css)
```

```python
# HTML Parser - Part 1
# Enter your code here. Read input from STDIN. Print output to STDOUT

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):

    def handle_starttag(self, tag, attrs):
        print(f"Start : {tag}")
        for attr in attrs:
            print(f"-> {attr[0]} > {attr[1]}")

    def handle_endtag(self, tag):
        print(f"End   : {tag}")


    def handle_startendtag(self, tag, attrs):
        print(f"Empty : {tag}")
        for attr in attrs:
            print(f"-> {attr[0]} > {attr[1]}")


N = int(input())
html_code = "\n".join([input() for _ in range(N)])

parser = MyHTMLParser()
parser.feed(html_code)
```

```python
# Validating Credit Card Numbers
# Enter your code here. Read input from STDIN. Print output to STDOUT

import re

def is_valid(card_number):

    pattern = r"^(?:4\d{3}|5\d{3}|6\d{3})(?:-?\d{4}){3}$"
    consecutive_repeats = r"(.)\1{3,}"

    if re.match(pattern, card_number):
        cleaned_number = card_number.replace("-", "")
        if re.search(consecutive_repeats, cleaned_number):
            return "Invalid"
        else:
            return "Valid"
    else:
        return "Invalid"


N = int(input())

for _ in range(N):
    print(is_valid(input()))
```

```python
# XML2 - Find the Maximum Depth
maxdepth = 0

def depth(elem, level):
    
    global maxdepth
    if level == maxdepth:
        maxdepth += 1
    for child in elem:
        depth(child, level + 1)
```

```python
# Complete the 'kangaroo' function below.
# The function is expected to return a STRING.
# The function accepts following parameters:

#  1. INTEGER x1
#  2. INTEGER v1
#  3. INTEGER x2
#  4. INTEGER v2

def kangaroo(x1, v1, x2, v2):

    if v1 == v2:
        return "YES" if x1 == x2 else "NO"
    
    n = (x2 - x1) / (v1 - v2)

    if n.is_integer() and n > 0:
        return "YES"
    else:
        return "NO"


if __name__ == '__main__':

    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    first_multiple_input = input().rstrip().split()

    x1 = int(first_multiple_input[0])
    v1 = int(first_multiple_input[1])
    x2 = int(first_multiple_input[2])
    v2 = int(first_multiple_input[3])

    result = kangaroo(x1, v1, x2, v2)
    fptr.write(result + '\n')
    fptr.close()
```

```python
# Recursive Digit Sum
#!/bin/python3



import math

import os

import random

import re

import sys



#

# Complete the 'superDigit' function below.

#

# The function is expected to return an INTEGER.

# The function accepts following parameters:

#  1. STRING n

#  2. INTEGER k

#



def superDigit(n, k):

    n = sum([int(c) for c in n]) * k

    while n >= 10:
        n = sum([int(c) for c in str(n)])

    return n



if __name__ == '__main__':

    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    first_multiple_input = input().rstrip().split()
    n = first_multiple_input[0]
    k = int(first_multiple_input[1])
    result = superDigit(n, k)
    fptr.write(str(result) + '\n')
    fptr.close()
```

```python
# Complete the 'insertionSort1' function below.

# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER_ARRAY arr


def insertionSort1(n, arr):

    # Write your code here

    if n == 1:
        print(arr)  

    num = arr[n-1]
    arr[n-1] = arr[n-2]

    for i in range(n-1, -1, -1):
        if num < arr[i-1] and i != 0:
            arr[i] = arr[i-1]
            print(" ".join(map(str, arr)))
        else:
            arr[i] = num
            print(" ".join(map(str, arr)))
            break


if __name__ == '__main__':

    n = int(input().strip())
    arr = list(map(int, input().rstrip().split()))
    insertionSort1(n, arr)
```

```python
# Insertion Sort - Part 2
#!/bin/python3

import math
import os
import random
import re
import sys


# Complete the 'insertionSort2' function below.

# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER_ARRAY arr

def insertionSort2(n, arr):

    for i in range(1, n):
        value_to_insert = arr[i]
        j = i - 1


        while (j >= 0) and (arr[j] > value_to_insert):
            arr[j + 1] = arr[j]
            j -= 1

        
        arr[j + 1] = value_to_insert
        print(' '.join(map(str, arr)))



if __name__ == '__main__':

    n = int(input().strip())
    arr = list(map(int, input().rstrip().split()))

    insertionSort2(n, arr)
```
